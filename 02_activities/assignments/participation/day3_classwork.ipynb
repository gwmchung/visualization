```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# plotting two y variables on the same axis
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot(x, y1)
ax.plot(x, y2)

# alternate way
# ax.plot(x, y1, y2)
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc = "lower right", 
          frameon = True, 
          fontsize = 12, 
          ncol = 2, 
          shadow = True)
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc = "lower left",   # loc refers to the corner of the legend box, where to anchor to
          bbox_to_anchor=(1, 0.5))  # first number if x axis location, second is y axis location; 0 - 1 is the % on the line
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "person 1")
ax.scatter(x, y2, label = "person 2")
ax.legend(loc = "lower right")
ax.text(30, 65, "This value is important!",
        ha = "center",
        color = "red",
        size = 20)
ax.grid()
```

```python
# demo the use of transform, which is what the first two numbers are relative to 
# https://matplotlib.org/stable/users/explain/artists/transforms_tutorial.html
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, "Data: (1,5)", transform = ax.transData)
ax.text(0.5, 0.1, "Axes: (0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, "Figure: (0.2, 0.2)", transform = fig.transFigure)
```

```python
# add an arrow to point to a specific point
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.annotate("This is important!", xy = (10, 95), xytext = (20, 94),
            #arrowprops = dict(facecolor = 'black', shrink = 1)) # 10, 95 is where arrow points to, 20, 94 is where text goes
            arrowprops = dict(arrowstyle = "wedge",
                              color = "hotpink"))
```

```python
# remove numbers on axes, remove tickmarks on y-axis
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# limit number of tickmarks on x-axis
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# specify the interval of the tickmarks
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# further customize tickmarks
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

plt.xticks(rotation = 45, ha = "right", fontsize = 12, font = "serif")
```

```python
# customize X axis with font dict
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

plt.xticks(rotation = 45, ha = "right", fontsize = 12, font = "serif")

font1 = {'family':'serif', 'color':'indigo', 'size':20}
plt.xlabel("Shiny New X Axis", fontdict=font1)
```
