```python
# Load packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# exercise to customize the plot
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize = (7, 3))
ax1.scatter(x1, y1)
ax2.bar(x2, y2, color = 'purple')

font1 = {'family':'serif', 'color':'indigo', 'size':20}

ax1.set_title('Plot1 title')
ax2.set_title('Plot2 title', fontdict=font1)
```

```python
# arrange multiple plots on a figure; denote position and how big each plot are (how many spaces it takes up) with list of list
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7, 4),
                                    layout = "constrained") # adding layout parameter enables everything to be fit in

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)
someaxes["ax1"].set_xlabel("A big label!", fontsize = 18)
someaxes["ax2"].set_xlabel("Another big label", fontsize = 18)
someaxes["ax3"].set_xlabel("Label2: 2 Fast 2 Furious", fontsize = 18)
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7, 3))

ax.bar(x, y1, 
       color = 'indigo')
ax.plot(x, y2,
        color = 'red')
```

```python
y2_sd =  np.std(y2)

fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2, 
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2)
```
