```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
#  Set a seed for reproduceable results
np.random.seed(613)
x = np.arange(50)       # create list of numbers from 0-49
y = np.random.randint(0, 100, 50)   # generate a random list of 50 numbers 
```

```python
# scatter plot 
fig, ax = plt.subplots(figsize = (5, 3)) 
ax.scatter(x, y)
```

```python
# line graph
fig, ax = plt.subplots(figsize = (5, 3))
#ax.bar(x, y)
ax.plot(x, y)
```

```python
# histogram
fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y)
```

```python
# Adding labels to the plot
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
#fig.tight_layout()
```

```python
# manipulate font in the graphs
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color' : 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2, loc = 'top')
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# manipulate marker and color
fig, ax = plt.subplots(figsize = (5, 3))

ax.scatter(x, y, marker = '*', 
           color = 'teal',
           )
```

```python
# manipulate properties of the line
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x,
        y,
        marker = '*',
        color = 'indigo',
        linestyle = '--',
        linewidth = '3')
```

```python
# practice exercise to customize the graph
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y,
        marker = '*',
        markersize = 15,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2, 
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'both', color = 'red', linestyle = 'dotted')
```
